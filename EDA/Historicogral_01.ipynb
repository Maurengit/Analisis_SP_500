**S&P 500**

El índice Standard & Poor's 500 (Standard & Poor's 500 Index), también conocido como S&P 500, es uno de los índices bursátiles más importantes de Estados Unidos. Al S&P 500 se lo considera el índice más representativo de la situación real del mercado.

```python
import pandas as pd
import yfinance as yf #yahoo finance
import matplotlib.pyplot as plt
import seaborn as sns

```

**Para obtener información histórica y análisis del índice S&P500**

```python
#Define un objeto ticker para el índice SP500 utilizando yfinance:
tickerSymbol = '^GSPC'
tickerData = yf.Ticker(tickerSymbol)
```

```python
#Obtén el histórico de precios del índice SP500 utilizando el método history de yfinance:
tickerDf = tickerData.history(period='1d', start='2000-1-1', end='2023-2-23')
```

```python
#Analizar los datos obtenidos con pandas, retorno diario del índice
tickerDf['Daily Return'] = tickerDf['Close'].pct_change()
```

```python
# Visualizar los datos, precio de cierre diario
plt.plot(tickerDf['Close'])
plt.show()
```
